### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo
```

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:-1]))
```

```python


chipo_order_price = chipo.groupby("order_id").sum()
chipo_order_price
```

```python
plt.scatter(x=chipo_order_price.item_price, y=chipo_order_price.quantity, c="green", edgecolors="black", norm="normalize")
plt.title("Number of items ordered per item price")
plt.xlabel("Order price")
plt.ylabel("items ordered")
plt.xlim([0, 250])
plt.ylim([0, 40])
```
