### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips
```

```python
x = sns.JointGrid(x="total_bill", y="tip", data=tips)
x = x.plot_joint(sns.scatterplot)
x.plot_marginals(sns.distplot, kde=False)
```
